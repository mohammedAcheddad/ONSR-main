```python
import pandas as pd
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
import random
import xgboost as xgb
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import *

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('UserTravelHistory/#2/data.csv')
df2 = pd.read_csv('UserTravelHistory/#2/dataOriginal.csv')

# Print the first few rows of the DataFrame
df.head()
```

```python
features = df.columns
feature = features.drop("target")
X = df[feature]
y= df['target']
```

```python
# split the data into train and test sets
x_train, x_test, y_train, y_test= train_test_split(X, y,
                                                   test_size= 0.2,
                                                   shuffle= True, #shuffle the data to avoid bias
                                                   random_state= 0)
print(f'training set size: {x_train.shape[0]} samples \ntest set size: {x_test.shape[0]} samples')
```

```python
import xgboost as xgb
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import roc_auc_score


params = { 'max_depth': [5,6,7],
           'learning_rate': [0.1],
           'n_estimators': [300,400,500,600],
           'colsample_bytree': [0.65,0.75]}
xgbr = xgb.XGBClassifier(use_label_encoder=False,verbosity=0)
clf = GridSearchCV(estimator=xgbr, 
                param_grid=params,
                   scoring='roc_auc', 
                   verbose=1)
clf.fit(x_train, y_train)
```

```python
xgbr = xgb.XGBClassifier(base_score=0.5, booster='gbtree', colsample_bylevel=1,
              colsample_bynode=1, colsample_bytree=0.65,
              enable_categorical=False, gamma=0, gpu_id=-1,
              importance_type=None, interaction_constraints='',
              learning_rate=0.1, max_delta_step=0, max_depth=5,
              min_child_weight=1, monotone_constraints='()',
              n_estimators=300, n_jobs=12, num_parallel_tree=1,
              predictor='auto', random_state=0, reg_alpha=0, reg_lambda=1,
              scale_pos_weight=1, subsample=1, tree_method='exact',
              use_label_encoder=False, validate_parameters=1, verbosity=0)
```

```python
xgbr.fit(x_train,y_train)
```

```python
buying_prob = []
Nbuying_prob = []
for index,offer in x_test.iterrows():
    offer = offer[feature].values.reshape(1, -1)
    pred = xgbr.predict(offer)
    proba = xgbr.predict_proba(offer)
    if pred == 1:
        buying_prob.append([offer,proba[0, 1]])
    else:
        Nbuying_prob.append([offer,proba[0, 0]])
    
buying_prob = sorted(buying_prob, key=lambda x: x[1], reverse=True)
Nbuying_prob = sorted(Nbuying_prob, key=lambda x: x[1])
sort_probs = buying_prob + Nbuying_prob
sort_probs
```
